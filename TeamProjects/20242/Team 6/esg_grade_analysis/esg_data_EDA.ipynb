```python
columns_to_keep = ['ticker', 'name', 'industry', 'environment_grade', 'social_grade', 'governance_grade', 'total_grade']
esg_cleaned = esg[columns_to_keep]
```

```python
esg_cleaned.dropna(subset=['industry', 'environment_grade', 'social_grade', 'governance_grade', 'total_grade'], inplace=True)

# Group the cleaned dataset by industry and aggregate company grades
grouped_data = esg_cleaned.groupby('industry').agg({
    'ticker': 'count',
    'environment_grade': 'first',  
    'social_grade': 'first',
    'governance_grade': 'first'
}).reset_index()

# Rename columns for better clarity
grouped_data.rename(columns={
    'ticker': 'company_count',
}, inplace=True)

# Display the grouped data
print(grouped_data)
```

```python
esg_tech = esg_cleaned[esg_cleaned['industry'] == "Technology"]
```

```python
esg_tech
```

```python
tech_lists = esg_cleaned[esg_cleaned['industry'] == "Technology"]['name'].tolist()
```

```python
# Group companies by environment grade
grouped_by_environment =esg_tech.groupby('environment_grade')['name'].apply(list).reset_index()
grouped_by_environment.rename(columns={'name': 'companies'}, inplace=True)

grouped_by_social = esg_tech.groupby('social_grade')['name'].apply(list).reset_index()
grouped_by_social.rename(columns={'name': 'companies'}, inplace=True)

# Group companies by governance grade
grouped_by_governance = esg_tech.groupby('governance_grade')['name'].apply(list).reset_index()
grouped_by_governance.rename(columns={'name': 'companies'}, inplace=True)

grouped_by_total = esg_tech.groupby('total_grade')['name'].apply(list).reset_index()
grouped_by_total.rename(columns={'name': 'companies'}, inplace=True)

# Display full lists of companies in each environment grade
def display_environment_grades(grouped_data):
    print("\nCompanies grouped by environment grade:")
    for _, row in grouped_data.iterrows():
        print(f"Environment Grade {row['environment_grade']}: {', '.join(row['companies'])}")

# Display full lists of companies in each social grade
def display_social_grades(grouped_data):
    print("Companies grouped by social grade:")
    for _, row in grouped_data.iterrows():
        print(f"Social Grade {row['social_grade']}: {', '.join(row['companies'])}")

# Display full lists of companies in each governance grade
def display_governance_grades(grouped_data):
    print("\nCompanies grouped by governance grade:")
    for _, row in grouped_data.iterrows():
        print(f"Governance Grade {row['governance_grade']}: {', '.join(row['companies'])}")

def display_total_grades(grouped_data):
    print("\nCompanies grouped by total grade:")
    for _, row in grouped_data.iterrows():
        print(f"Total Grade {row['total_grade']}: {', '.join(row['companies'])}")
```

```python
# Display the grouped results
display_environment_grades(grouped_by_environment)
print("\n-----------------------------------------------------------------\n")
display_social_grades(grouped_by_social)
print("\n-----------------------------------------------------------------\n")
display_governance_grades(grouped_by_governance)
print("\n-----------------------------------------------------------------\n")
display_total_grades(grouped_by_total)
```
